```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# Load the Excel file
df = pd.read_excel('nominal_expenditures2.xlsx', header = 4)
pd.set_option('display.max_columns', None)
df.head(26)
df[df.Year > 2009]
df=df[df.Year > 2009]
df[df.Year < 2021]
df=df[df.Year < 2021]
df
```

```python
#Bar Plot displaying Total FAFH in Millions of U.S. Dollars from 2010-2020
sns.barplot(x="Year", y= "Total FAFH", data=df)
plt.title('Total FAFH in Millions of U.S. Dollars from 2010-2020')
plt.savefig('totalfafh_bar.png')
plt.show()
#df.round(1)
```

```python
# Load the Excel file
df = pd.read_excel('monthly_sales.xlsx')
twenty = df.loc[(df["Year"]==2020)]
twenty_one= df.loc[(df["Year"]==2021)]
twenty_two= df.loc[(df["Year"]==2022)]

#reverse
twenty = twenty.iloc[::-1]
twenty_one = twenty_one.iloc[::-1]
twenty_two = twenty_two.iloc[::-1]

# Remove rows with any missing values
df.dropna(inplace=True)
df = df[df['Year'] >= 2020]
pd.set_option('display.max_columns', None)
twenty
```

```python
#creating variables for lineplot

x_axis = twenty["Month"]
y_twenty = twenty["FAFH"]
y_twenty_one = twenty_one["FAFH"]
y_twenty_two = twenty_two["FAFH"]

#lineplot
plt.figure(figsize=(10, 6))

#plot data for 2020, 2021, 2022
plt.plot(x_axis, y_twenty, marker= "v", label='2020')
plt.plot(x_axis, y_twenty_one, marker = "o", label='2021')
plt.plot(x_axis, y_twenty_two, marker = "d", label='2022')

# plot labels and title
plt.xlabel('Month')
plt.ylabel('FAFH')
plt.title('FAFH: 2020-2022')

# Show the plot
plt.legend()
plt.xticks(rotation=90)
plt.show()
```
