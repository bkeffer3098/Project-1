```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Import data
state_data = Path("../Resources/state_sales.xlsx")

#Read the data
read_state = pd.read_excel(state_data)
read_state
```

```python
#Slim down the data to the states I need
populous_states = read_state.loc[(read_state['State'] == "California") | (read_state['State'] == "New York") |
                            (read_state['State'] == "Florida") | (read_state['State'] == "Ohio") |
                            (read_state['State'] == "Texas"),:]
populous_states
```

```python
#Slim down to the last 10 years. 
populous_states = populous_states.loc[(populous_states['Year'] >= 2010) ,:]

populous_states.head()
```

```python
california_df = populous_states.loc[(populous_states['State'] == "California") ,:]
newyork_df = populous_states.loc[(populous_states['State'] == "New York") ,:]
texas_df = populous_states.loc[(populous_states['State'] == "Texas") ,:]
florida_df = populous_states.loc[(populous_states['State'] == "Florida") ,:]
ohio_df = populous_states.loc[(populous_states['State'] == "Ohio") ,:]
```

```python
#line plot that shows the FAH spending per state
#each state has a different x and y 

x_cal = california_df["Year"]
y_cal = california_df["FAFH"]

x_ny = newyork_df["Year"]
y_ny = newyork_df["FAFH"]

x_tex = texas_df["Year"]
y_tex = texas_df["FAFH"]

x_flo = florida_df["Year"]
y_flo = florida_df["FAFH"]

x_ohio = ohio_df["Year"]
y_ohio = ohio_df["FAFH"]

```

```python
#plotting

plt.plot(x_cal, y_cal,marker="+",color="red", linewidth=1, label="California")
plt.plot(x_ny, y_ny,marker="+",color="blue", linewidth=1, label="New York")
plt.plot(x_tex, y_tex,marker="+",color="green", linewidth=1, label="Texas")
plt.plot(x_flo, y_flo,marker="+",color="black", linewidth=1, label="Florida")
plt.plot(x_ohio, y_ohio,marker="+",color="pink", linewidth=1, label="Ohio")


plt.xlabel("Years")
plt.ylabel("Sales (in Millions of Dollars)")
plt.title("Food Away From Home in Most Populous States")
plt.legend(loc="best", fontsize="8")

plt.savefig("../output_data/MostPopulousStatesFAFH.png")
plt.show()
```
