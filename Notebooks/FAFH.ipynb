```python
# Load the Excel file
df = pd.read_excel('../Resources/monthly_sales.xlsx')
twenty = df.loc[(df["Year"]==2020)]
twenty_one= df.loc[(df["Year"]==2021)]
twenty_two= df.loc[(df["Year"]==2022)]

#reverse
twenty = twenty.iloc[::-1]
twenty_one = twenty_one.iloc[::-1]
twenty_two = twenty_two.iloc[::-1]

# Remove rows with any missing values
df.dropna(inplace=True)
df = df[df['Year'] >= 2020]
pd.set_option('display.max_columns', None)
twenty
```

```python
#creating variables for lineplot

x_axis = twenty["Month"]
y_twenty = twenty["FAFH"]
y_twenty_one = twenty_one["FAFH"]
y_twenty_two = twenty_two["FAFH"]

#lineplot
plt.figure(figsize=(10, 6))

#plot data for 2020, 2021, 2022
plt.plot(x_axis, y_twenty, marker= "v", label='2020')
plt.plot(x_axis, y_twenty_one, marker = "o", label='2021')
plt.plot(x_axis, y_twenty_two, marker = "d", label='2022')

# plot labels and title
plt.xlabel('Month')
plt.ylabel('FAFH')
plt.title('FAFH: 2020-2022')

# Show the plot
plt.savefig('../output_data/fafh_2020_2022_linechart.png')
plt.legend()
plt.xticks(rotation=90)
plt.show()
```
